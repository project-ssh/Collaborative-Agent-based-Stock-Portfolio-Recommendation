```python
data=pd.read_csv("/content/drive/MyDrive/Dataset/Stock_Data.csv")
```

```python
all_companies = data['Ticker'].unique()
```

```python
companies =[]
for i in range(len(all_companies)):
  if(i>556 and i<=656):
    companies.append(all_companies[i])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.layers import Layer, Input, Bidirectional, LSTM, Dense
from tensorflow.keras.models import Model

# Define the sequence length for the model
seq_length = 4  # Example sequence length

# Define the Attention layer
class Attention(Layer):
    def _init_(self, **kwargs):
        super(Attention, self)._init_(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name='attention_weight', shape=(input_shape[-1], input_shape[-1]), initializer='random_normal', trainable=True)
        self.b = self.add_weight(name='attention_bias', shape=(input_shape[-1],), initializer='random_normal', trainable=True)
        super(Attention, self).build(input_shape)

    def call(self, x):
        e = tf.keras.backend.tanh(tf.keras.backend.dot(x, self.W) + self.b)
        a = tf.keras.backend.softmax(e, axis=1)
        output = x * a
        return tf.keras.backend.sum(output, axis=1)

# Define the BiLSTM model with attention
def build_bilstm_attention_model(seq_length, feature_dim):
    inputs = Input(shape=(seq_length, feature_dim))
    lstm_out = Bidirectional(LSTM(50, return_sequences=True))(inputs)
    lstm_out = Bidirectional(LSTM(50, return_sequences=True))(lstm_out)
    attention_out = Attention()(lstm_out)
    outputs = Dense(1)(attention_out)
    model = Model(inputs, outputs)
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Function to create sequences
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:i+seq_length]
        y = data[i+seq_length]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# DataFrame to store the predicted values
all_predictions = pd.DataFrame()
count = 1
# Iterate over each ticker in the dataset
for ticker in companies:
    print(f"Predicting for {count}){ticker}...")
    count+=1
    # Filter data for the current ticker
    stock_data = data[data['Ticker'] == ticker]
    stock_data.set_index('Date', inplace=True)
    stock_data.index = pd.to_datetime(stock_data.index)

    # Use the 'Close' price for forecasting
    df = stock_data[['Close']]

    # Fill missing values (if any)
    df.fillna(method='ffill', inplace=True)

    # Normalize the data
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df.values)

    # Split the data into training and test sets
    train_size = int(len(scaled_data) * 0.8)
    train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]

    # Create sequences
    X_train, y_train = create_sequences(train_data, seq_length)
    X_test, y_test = create_sequences(test_data, seq_length)

    if len(X_train) == 0 or len(X_test) == 0:
        print(f"Not enough data to create sequences for {ticker}. Skipping.")
        continue

    # Build and train the model
    feature_dim = X_train.shape[2]
    model = build_bilstm_attention_model(seq_length, feature_dim)
    model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=0)

    # Make predictions
    y_pred = model.predict(X_test)
    y_pred = scaler.inverse_transform(y_pred)
    y_test = scaler.inverse_transform(y_test)

    # Create a DataFrame for the predicted prices with the ticker
    predicted_df = pd.DataFrame(y_pred, index=df.index[-len(y_pred):], columns=['Predicted_Close'])
    predicted_df['Ticker'] = ticker

    # Append to the all_predictions DataFrame
    all_predictions = pd.concat([all_predictions, predicted_df])

    # Plotting the results for each ticker
    plt.figure(figsize=(14, 5))
    plt.plot(df.index[-len(y_test):], y_test, color='blue', label='Actual')
    plt.plot(predicted_df.index, predicted_df['Predicted_Close'], color='red', label='Predicted')
    plt.title(f'Time Series Forecasting with BiLSTM for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.show()

# Display the combined predictions DataFrame
print(all_predictions)

# Optionally save to a CSV file
all_predictions.to_csv('all_predictions.csv', index=True)
```
